# **DATA CLEANING:**

```python
#check for any null values
df.isnull().sum()
```

```python
#calculating the percentage of columns
df.isnull().sum() * 100 / len(df)
```

```python
#now remove the columns with 70% + percentage
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)

```

```python
#remove the null values from remaining column
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.describe()
```

# **EDA (Exploratory data analysis):**

```python
#check the columns
df.columns
```

**Gender:**

```python

gen= sns.countplot(data = df, x = 'Gender')

for bars in gen.containers:
    gen.bar_label(bars)
```

```python
gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y= 'Amount' ,data =  gen)
```

*from the above graph we can estimate that `women` purchased more than men.*

**Marriage Status:** 

```python
mrg= sns.countplot(data = df, x = 'Marital_Status')

for bars in mrg.containers:
    mrg.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

*From above graph we see that `married women` bought more than `married men`*

**Age:**

```python
age = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in age.containers:
    age.bar_label(bars)
```

```python
age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y= 'Amount' ,data = age)
 
```

*Above graph shows that people with age group from `26-35`  have purchased more*

**Product Category:**

```python

plt.figure(figsize=(12, 6))   

pro = sns.countplot(data=df, x='Product_Category')
for bars in pro.containers:
    pro.bar_label(bars)

plt.xticks(rotation=90, ha='right')   
plt.show()

```

```python
pro = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(12, 6))   
sns.barplot(data = pro, x = 'Product_Category',y= 'Amount')
plt.xticks(rotation=90, ha='right')   
plt.show()
```

*Above graphs shows that most of the sold products are from `Food`, `Clothing` and `Electronics and gadgets`*

**Occupation:**

```python

plt.figure(figsize=(12, 6))   

occp = sns.countplot(data=df, x='Occupation')
for bars in occp.containers:
    occp.bar_label(bars)

plt.xticks(rotation=90, ha='right')   
plt.show()

```

```python
occp = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.figure(figsize=(12, 6))   
sns.barplot(data = occp, x = 'Occupation',y= 'Amount')
plt.xticks(rotation=90, ha='right')   
plt.show()
```

*Above graph shows that most of the buyers are from in `IT sector`, `Healthcare` and `Aviation`*
